## Alpha Channel
- An alpha value of zero represents full transparency
- A value of 255 represents a fully opaque pixel

$$
    I = \alpha \times F + (1 - \alpha) \times B, \quad \alpha \in [0,1]
$$

```python
plt.imshow(im[:,:,-1], cmap="gray")
plt.title("alpha channel")
```
